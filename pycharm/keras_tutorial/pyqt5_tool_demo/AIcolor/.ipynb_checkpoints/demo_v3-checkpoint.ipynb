### 图像识别的基本流程
- 获取数据
尽可能接近工程实际的数据，更具有泛化性
保留原始数据，进行备份（重要）
编写基本文档：如每种类别的数量，中英文名称，基本的数据格式等，写在一个 Readme.md 文档中
- 检查数据
对数据进行类别分类，检查样本中的异常数据（如图片的格式是否一致），不能对数据进行任何操作
使用对比软件对进行分类的数据和原始数据进行对比，防止错误的操作，没有人愿意去做哪些意义的工作，将数据错误的概率降到最低

### 数据处理

```python
''' 
# 初始时用来创建文件夹的

def create_dir(base_dir, class_names):
    """
    base_path: 放置文件的路径
    class_names: 类别的名字
    """
    base_dir = os.path.join(base_dir, "img_data")
    
    if os.path.exists(base_dir) is False:
        os.mkdir(base_dir)
        # 创建 train test validation 文件夹，下面放每个类别的样本
        train_dir = os.path.join(base_dir, "train")
        test_dir = os.path.join(base_dir, "test")
        validation_dir = os.path.join(base_dir, "validation")

        os.mkdir(train_dir)
        os.mkdir(test_dir)
        os.mkdir(validation_dir)

        # 创建不同类的文件夹
        for class_name in class_names:
            os.mkdir(os.path.join(train_dir, class_name))
            os.mkdir(os.path.join(test_dir, class_name))
            os.mkdir(os.path.join(validation_dir, class_name))
    else:
        print("filepath already exist")

base_dir = "./"
class_names = ["SAPwood", "Mineral", "brown"] # 到时根据需求更改
create_dir(base_dir, class_names)
'''


path_train = 'f:/jupyter/wood_detection/img_data/train/'
path_test = 'f:/jupyter/wood_detection/img_data/test/'
path_val = 'f:/jupyter/wood_detection/img_data/validation/'


def load_img_label(path):
    target_size = (248, 248)
    files = os.listdir(path)
    cate = []
    img_path_list = []
    labels = []
    imgs = []
    for f in files:
        print(f)
        cate.append(f)
        cate_path = path + '/' + f
        img_list = os.listdir(cate_path)
        for img_each in img_list:
            img_path = cate_path + '/' + img_each
            #print(img_path)
            img_path_list.append(img_path)
            img = image.load_img(img_path, target_size=target_size)
            img_array = image.img_to_array(img)
            imgs.append(img_array)
            labels.append(cate.index(img_path.split('/')[-2]))

    data = np.array(imgs)
    labels = np.array(labels)
    labels = np_utils.to_categorical(labels, 3)
    return data, labels
```

```python
# 把图片读取到计算机内存中，方便 GPU 加速
train_data, train_label = load_img_label(path_train)
#test_data, test_label = load_img_label(path_test)
val_data, val_label = load_img_label(path_val)

```

### 数据增强

```python
train_datagen = ImageDataGenerator(rescale=1./255,  # 图像进行缩放
                                   shear_range=0.2,  # 随机错切变换的角度
                                   zoom_range=0.2,  # 图像随机缩放的范围。
                                   horizontal_flip=True)  # 随机将一半图像水平翻转

validation_datagen = ImageDataGenerator(rescale=1./255)  # 不能对验证数据集进行数据增强

test_datagen = ImageDataGenerator(rescale=1./255)

 
train_generator = train_datagen.flow(x=train_data, y=train_label, batch_size=32) 
validation_generator = validation_datagen.flow(x=val_data, y=val_label, batch_size=32) 
```

### 建立模型函数

```python
#建立模型 
# 一些参数设置
input_shape = (248,248, 3)

def data_generator_model_demo(class_num):
    model = Sequential()
    model.add(Conv2D(32, (5, 5), activation='relu', input_shape=input_shape, padding='same', strides=2)) # 输入大小
    model.add(MaxPooling2D((2, 2)))
    
    model.add(Conv2D(32, (3, 3), activation='relu', input_shape=input_shape, padding='same')) # 输入大小
    model.add(MaxPooling2D((2, 2)))
    
    model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
    model.add(MaxPooling2D((2, 2)))
    
    model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
    model.add(MaxPooling2D((2, 2)))

    model.add(Conv2D(128, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))
    
    model.add(Flatten())
    
    model.add(Dropout(0.5)) # 增加了 dropout 层
    model.add(Dense(1000, activation='relu'))
    model.add(Dense(class_num, activation='softmax')) # 多分类，最后一层使用 softmax
    return model
```

```python
# 创建回调函数
callbacks_list = [EarlyStopping(monitor='acc',  # 监控模型的验证精度
                                patience=20,
                                verbose=2),  # 若精度训练轮数大于此数不再改善，则中断训练
                  ModelCheckpoint(  # 每轮过后保存当前权重
                      filepath="model_path/{}.h5".format(time()),  # 目标模型文件的保存路径
                      monitor='val_loss',  # 若 val_loss 没有改善，那么不需要覆盖模型文件，这样保存训练过程中的最佳模型
                      save_best_only=True),
                  ReduceLROnPlateau( 
                      monitor='val_loss', # 监控模型的验证损失
                      factor=0.1, # 触发时将学习率除以 10
                      patience=2) # 如果验证损失在 10 轮内都没有改善，那么就触发这个回调函数
                 ]
```

### 创建模型并编译

```python

model = data_generator_model_demo(class_num=3)

model.compile(optimizer=RMSprop(lr=0.0004), # 在本书中说过，直接使用 RmSprop 是稳妥的。
              loss='categorical_crossentropy', 
              metrics=['accuracy']) 

model.summary()
```

```python
STEP_SIZE_TRAIN =train_generator.n // train_generator.batch_size
STEP_SIZE_VALID = validation_generator.n // validation_generator.batch_size
STEP_SIZE_TRAIN, STEP_SIZE_VALID
```

```python
history = model.fit_generator(generator=train_generator,
                              steps_per_epoch=128,
                              epochs=120, 
                              validation_data=validation_generator, 
                              validation_steps=64,
                              callbacks=callbacks_list
                            )
```

```python
model.summary()
```

### 可视化精度和损失

```python
import matplotlib.pyplot as plt
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(acc) + 1)

plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()

plt.figure()
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()
plt.show()
```
